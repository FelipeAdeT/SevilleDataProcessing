```python
nextbutton = browser.find_element_by_xpath("//li[@class='pager__item pager__item--next']/a")
nextbutton.click()
```

```python
while nextbutton:
    namelist= browser.find_elements_by_class_name("field-content")
    for result in namelist:
        nombres.append(result.text)
    try:
        nextbutton = browser.find_element_by_xpath("//li[@class='pager__item pager__item--next']/a")
        nextbutton.click()
    except:
        break
```

```python
nombres_cleaning = []

for name in nombres:
    if name.find('.') != -1:
        a=name.find('.')
        nom = name[:a]
        nombres_cleaning.append(nom)
    else:
        nombres_cleaning.append(name)
```

```python
nombres_cleaning
```

```python
fnames=[]
lnames = []

for name in nombres_cleaning:
    comma= name.find(',')
    lname = name[:comma]
    fname = name[comma+1:]
    fnames.append(fname)
    lnames.append(lname)
```

```python
# Split into names and last names
# Unique lists
# Fuzzy matching OR generate pseudo-names like t can = th, b=v, v=u, s=f=ff=ss, l=ll, p=pp
```

```python
unique_fnames= []

for name in fnames: 
        # check if exists in unique_list or not 
    if name not in unique_fnames: 
        unique_fnames.append(name)

print(unique_fnames)
len(unique_fnames)
```

```python
#Cleaning: tokenize? remove de, la, numbers? remove accents? remove fray, 
# perhaps the easiest option is to include words that could be in real names
#Need to clean out ()
# Further cleaning: remove accents, lowercase

unique_fnames2 =[]
templist=[]

for name in unique_fnames:
    a= name.split()
    for split in a:
        templist.append(split)

for name in templist:
    name= name.replace(')','')
    name=name.replace('(','')
    name=name.replace('\'','')
    name= name.lower()
    name=name.replace('í','i')
    name=name.replace('ó','o')
    name=name.replace('ú','u')
    name=name.replace('ü','u')
    name=name.replace('á','a')
    name=name.replace('é','e')
    name=name.replace('ì','i')
    name=name.replace('ò','o')
    name=name.replace('ù','u')
    name=name.replace('à','a')
    name=name.replace('è','e')
    name=name.replace('î','i')
    name=name.replace('ô','o')
    name=name.replace('û','u')
    name=name.replace('â','a')
    name=name.replace('ê','e')
    if name not in unique_fnames2: 
        unique_fnames2.append(name)
        
print(unique_fnames2)
len(unique_fnames2)
```
